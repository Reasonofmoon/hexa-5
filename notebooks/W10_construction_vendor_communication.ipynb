## Step 2: 이슈유형별 외주업체 공문 Gemini3종생성→ZIP

```python
import pandas as pd, numpy as np
np.random.seed(42)
# 공정별 샘플 데이터
df=pd.DataFrame({'공종':[f'공종{i+1}' for i in range(10)],
    '예산(만원)':np.random.randint(500,5000,10),
    '실적(만원)':np.random.randint(400,5500,10),
    '진행률':np.random.uniform(30,100,10).round(1)})
# 이상치 보장 (W09용)
df.loc[2,'진행률']=5.0; df.loc[7,'진행률']=8.0
df['초과율']=((df['실적(만원)']-df['예산(만원)'])/df['예산(만원)']*100).round(1)
print(df.to_string(index=False))
```

## Step 3: AI 분석

```python
p=f"""건설업 AI 전환 전문 컨설턴트로서 이슈유형별 외주업체 공문 Gemini3종생성→ZIP를 수행하세요.
현장: {{INFO['name']}} {{INFO['site']}} ({{INFO['type']}})
인사이트 3가지 + 즉시 실행 액션. 마크다운."""
resp=model.generate_content(p); print(resp.text)
```
